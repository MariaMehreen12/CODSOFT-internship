```python
df["Class"].unique()
```

```python
fraud=df[df["Class"]==1]
Not_fraud=df[df["Class"]==0]
```

```python
Not_fraud=Not_fraud.sample(n=492)
```

```python
New_value=pd.concat([fraud,Not_fraud],axis=0)
New_value.head(10)
```

```python
corr_matrix=New_value.corr(method="pearson")
plt.figure(figsize=(25,25))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
```

```python
selected_features=New_value.drop("Class",axis='columns')
selected_features.head()
```

```python
A=New_value["Class"]
A.head()
```

```python
scaler=StandardScaler()
scaler.fit(selected_features)
selected_features=scaler.transform(selected_features)
```

```python
x_train,x_test,y_train,y_test=train_test_split(selected_features,A,test_size=0.2,stratify=A,random_state=2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
accuracy=accuracy_score(y_test,y_predict)
print("The Accuracy For the LogisticRegression is",int(accuracy*100),"%")
```

```python
from sklearn.tree import DecisionTreeClassifier
ab=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
x_train_prediction=ab.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print('accuracy for training data is',training_data_accuracy)
```

```python
x_test_prediction=ab.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('accuracy for test data is',test_data_accuracy)
```
